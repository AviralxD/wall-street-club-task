```python
import datetime as dt
import talib as ta
import numpy as np
import plotly.express as px
import pandas as pd
```

```python
adani = pd.read_csv('wall-street-data/ADANIENT.NS.csv')
axisbank = pd.read_csv('wall-street-data/AXISBANK.NS.csv')
bhartiartl = pd.read_csv('wall-street-data/BHARTIARTL.NS.csv')
cipla = pd.read_csv('wall-street-data/CIPLA.NS.csv')
hindalco = pd.read_csv('wall-street-data/HINDALCO.NS.csv')
hindunilvr = pd.read_csv('wall-street-data/HINDUNILVR.NS.csv')
infy = pd.read_csv('wall-street-data/INFY.NS.csv')
itc = pd.read_csv('wall-street-data/ITC.NS.csv')
maruti = pd.read_csv('wall-street-data/MARUTI.NS.csv')
sbin = pd.read_csv('wall-street-data/SBIN.NS.csv')
sunpharma = pd.read_csv('wall-street-data/SUNPHARMA.NS.csv')
tatasteel = pd.read_csv('wall-street-data/TATASTEEL.NS.csv')
wipro = pd.read_csv('wall-street-data/WIPRO.NS.csv')

stocks = [adani, axisbank, bhartiartl, cipla, hindalco, hindunilvr, infy, itc, maruti, sbin, sunpharma, tatasteel, wipro]
```

```python
# Calculating RSI, MACD, and Bollinger Bands (with triple exponential moving average) for all stocks
for stock in stocks:
  stock['RSI'] = ta.RSI(stock['Close']) 
  stock['MACD'], stock['MACD_Signal'], stock['MACD_Hist'] = ta.MACD(stock['Close'])
```

```python
#Graphing RSI for all stocks
for stock in stocks: 
  fig_RSI = px.line(stock, x = "Date", y = "RSI")
  fig_RSI.add_hline(y=70, line_dash = "dash", line_color = "red")
  fig_RSI.add_hline(y=30, line_dash = "dash", line_color = "green")
  fig_RSI.show()
```
